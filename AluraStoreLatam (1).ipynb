### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
facturacion_tienda1 = (tienda["Precio"] + tienda["Costo de envío"]).sum()
facturacion_tienda2 = (tienda2["Precio"] + tienda2["Costo de envío"]).sum()
facturacion_tienda3 = (tienda3["Precio"] + tienda3["Costo de envío"]).sum()
facturacion_tienda4 = (tienda4["Precio"] + tienda4["Costo de envío"]).sum()

print("Facturación Tienda 1:", facturacion_tienda1)
print("Facturación Tienda 2:", facturacion_tienda2)
print("Facturación Tienda 3:", facturacion_tienda3)
print("Facturación Tienda 4:", facturacion_tienda4)
```

```python
print("Columnas Tienda 1:", tienda.columns)
print("Columnas Tienda 2:", tienda2.columns)
print("Columnas Tienda 3:", tienda3.columns)
print("Columnas Tienda 4:", tienda4.columns)
```

```python
facturacion1 = (tienda["Precio"] + tienda["Costo de envío"]).sum()
facturacion2 = (tienda2["Precio"] + tienda2["Costo de envío"]).sum()
facturacion3 = (tienda3["Precio"] + tienda3["Costo de envío"]).sum()
facturacion4 = (tienda4["Precio"] + tienda4["Costo de envío"]).sum()

import pandas as pd

df_facturacion = pd.DataFrame({"Tienda": ["Tienda 1","Tienda 2","Tienda 3","Tienda 4"],"Facturación": [facturacion1, facturacion2, facturacion3, facturacion4]})
df_facturacion.plot(kind='bar', x='Tienda', figsize=(10,6), color='purple')
```

# 2. Ventas por categoría

```python
ventas_categoria1 = tienda["Categoría del Producto"].value_counts()
print(ventas_categoria1)
```

```python
df_categoria = tienda.groupby('Categoría del Producto')[['Precio']].count()
df_categoria.plot(kind='pie',y='Precio',figsize=(8,8))
```

```python
porcentaje_categoria = tienda["Categoría del Producto"].value_counts(normalize=True).to_frame()

porcentaje_categoria.plot(kind='bar',figsize=(12,8),color='green',xlabel='Categorías',ylabel='Porcentaje')
```

```python
ventas_categoria3 = tienda3["Categoría del Producto"].value_counts()
print(ventas_categoria3)
```

```python
ventas_categoria4 = tienda4["Categoría del Producto"].value_counts()
print(ventas_categoria4)
```

# 3. Calificación promedio de la tienda


```python
calif1 = tienda["Calificación"].mean()
calif2 = tienda2["Calificación"].mean()
calif3 = tienda3["Calificación"].mean()
calif4 = tienda4["Calificación"].mean()

print("Calificación promedio Tienda 1:", calif1)
print("Calificación promedio Tienda 2:", calif2)
print("Calificación promedio Tienda 3:", calif3)
print("Calificación promedio Tienda 4:", calif4)
```

# 4. Productos más y menos vendidos

```python
mas_vendido1 = tienda["Producto"].value_counts().idxmax()
menos_vendido1 = tienda["Producto"].value_counts().idxmin()

print("Producto más vendido Tienda 1:", mas_vendido1)
print("Producto menos vendido Tienda 1:", menos_vendido1)
```

```python
mas_vendido2 = tienda2["Producto"].value_counts().idxmax()
menos_vendido2 = tienda2["Producto"].value_counts().idxmin()

print("Producto más vendido Tienda 2:", mas_vendido2)
print("Producto menos vendido Tienda 2:", menos_vendido2)
```

```python
mas_vendido3 = tienda3["Producto"].value_counts().idxmax()
menos_vendido3 = tienda3["Producto"].value_counts().idxmin()

print("Producto más vendido Tienda 3:", mas_vendido3)
print("Producto menos vendido Tienda 3:", menos_vendido3)
```

```python
mas_vendido4 = tienda4["Producto"].value_counts().idxmax()
menos_vendido4 = tienda4["Producto"].value_counts().idxmin()

print("Producto más vendido Tienda 4:", mas_vendido4)
print("Producto menos vendido Tienda 4:", menos_vendido4)
```

# 5. Envío promedio por tienda

```python
envio1 = tienda["Costo de envío"].mean()
envio2 = tienda2["Costo de envío"].mean()
envio3 = tienda3["Costo de envío"].mean()
envio4 = tienda4["Costo de envío"].mean()

print("Envío promedio Tienda 1:", envio1)
print("Envío promedio Tienda 2:", envio2)
print("Envío promedio Tienda 3:", envio3)
print("Envío promedio Tienda 4:", envio4)
```
